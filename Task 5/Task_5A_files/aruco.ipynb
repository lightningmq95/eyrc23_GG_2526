```python
import cv2
import cv2.aruco as aruco
```

```python
VideoCap = True
cap = cv2.VideoCapture(0)
```

```python
def findAruco(img, marker_size = 6, total_markers = 250, draw = True):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    arucoDict = cv2.aruco.getPredefinedDictionary(cv2.aruco.DICT_4X4_250)
    arucoParam = aruco.DetectorParameters()
    bbox, ids, _ = cv2.aruco.detectMarkers(gray, arucoDict, parameters=arucoParam)
    print("Detected IDs", ids)
    if draw:
        if ids is not None:
            for i in range(len(ids)):
                c = tuple(map(int, bbox[i][0][0]))  # Convert the coordinates to integer
                cv2.putText(img, str(ids[i]), c, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 255, 0), 2, cv2.LINE_AA)
        cv2.aruco.drawDetectedMarkers(img, bbox)
    return bbox, ids
```

```python
while True:
    if VideoCap: _, img = cap.read()
    else:
        img = cv2.imread('Images/flex5.jpg')
        # img = cv2.resize(img, (0,0), fx = 0.4, fy= 0.4)
    bbox, ids = findAruco(img)
    if cv2.waitKey(1) == 113:
        break
    cv2.imshow('img', img)
```
